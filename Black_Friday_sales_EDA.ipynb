## Column Description
It is a description of the columns in the dataset, providing a understanding of each column:


- **User_ID**: Unique identifier for each user in the dataset.

- **Product_ID**: Identifier for each product purchased by the customer.

- **Gender**: Gender of the user, typically represented as "M" for male and "F" for female.

- **Age**: Age group of the user, typically provided in ranges such as "0-17", "18-25", "26-35", etc.

- **Occupation**: Code representing the occupation of the user. Each number corresponds to a specific occupation category.

- **City_Category**: Code ('A', 'B', 'C') indicating the size and type of city the customer resides in like Tier-1, Tier-2, Tier-3 cities.

- **Stay_In_Current_City_Years**: Duration (in years) the customer has lived in the current city, with '4+' representing four or more years.

- **Marital_Status**: Indicates if the customer is married (1) or unmarried (0).

- **Product_Category_1**: Primary category code for the product type (e.g., 1 = Electronics, 3 = Clothing).

- **Product_Category_2**: Secondary, optional sub-category for further product details (e.g., 6 = Accessories for Electronics).

- **Product_Category_3**: Tertiary, optional sub-category adding finer details (e.g., 14 = Premium Accessories).

- **Purchase**: Purchase amount spent by the user on the specific product, represented as a numeric value.

Q1: What is the average purchase amount based on gender and age groups?

```python
gender_amount = data.groupby('Gender').agg({'Amount':'mean'})
print(gender_amount)
```

Q2: Create a boxplot and What do you observe and describe the distribution of the Purchase column based on age group.

```python
sns.boxplot(x='Gender', y='Amount', data=data)
```

Q3: What are the total purchase amounts across different Cities? Use a bar plot to represent your findings.

```python
city_purchase = data.groupby('City_Category').agg({'Amount':'sum'}).reset_index()
sns.barplot(x='City_Category', y='Amount', data=city_purchase)
```

Q4: Which age group has the 2nd highest average purchase amount by gender? Present it in the visual format.

```python
# Step 1: Group by Age & Gender and compute average purchase
age_gender_avg = data.groupby(['Age', 'Gender'])['Amount'].mean().reset_index()

# Step 2: Sort by average purchase (highest  lowest)
sorted_age_gender = age_gender_avg.sort_values('Amount', ascending=False)

# Step 3: Select the 2nd highest AgeGender combination
second_highest = sorted_age_gender.iloc[1]

# Step 4: Filter only that row
target = age_gender_avg[
    (age_gender_avg['Age'] == second_highest['Age']) &
    (age_gender_avg['Gender'] == second_highest['Gender'])
]

# Step 5: Visualize
plt.figure(figsize=(6,4))
sns.barplot(x='Age', y='Amount', hue='Gender', data=target)
plt.title("2nd Highest Average Purchase Amount by Age & Gender")
plt.ylabel("Average Purchase Amount")
plt.show()
```

Q7. How does the average purchase amount depict in comparison with Product_Category_1 & Product_Category_2 ? Create a distribution plot to represent this.

```python
avg_pc1 = data.groupby('Product_Category_1')['Amount'].mean()
avg_pc2 = data.groupby('Product_Category_2')['Amount'].mean()
plt.figure(figsize=(12,5))
sns.histplot(avg_pc1, kde=True)
plt.figure(figsize=(12,5))
sns.histplot(avg_pc2, kde=True)
```

Q8: Identify the top 5 products with the highest purchase amounts.

```python
top_5_products = data.groupby('Product_ID')['Amount'].sum()
print(top_5_products.nlargest(5))
```

Q9: Which age group has the highest average purchase amount?

```python
age_amount = data.groupby('Age').agg({'Amount': 'mean'})
print(age_amount['Amount'].nlargest(1))
```

Q10: For each occupation, calculate the percentage of top 5 total purchases, and visualise it.

```python
occupation_amount = data.groupby('Occupation')['Amount'].sum().reset_index()
top5 = occupation_amount.sort_values('Amount', ascending=False).head(5)
top5['Percent'] = (top5['Amount'] / top5['Amount'].sum()) * 100
plt.figure(figsize=(4,4))
plt.pie(
    top5['Percent'],
    labels=top5['Occupation'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Top 5 Occupations by Total Purchase Percentage")
plt.show()
```

Q13: What is the maximum purchase amount recorded for each Product Category?

```python
cat_amount = data.groupby('Product_Category_1').agg({'Amount':'max'})
max_cat_amount = cat_amount.sort_values('Amount', ascending=False)
max_cat_amount
```

Q14: Investigate how many unique products were purchased by each User_ID and visualize the results.

```python
user_product_count = data.groupby('User_ID')['Product_ID'].nunique().reset_index()
user_product_count.rename(columns={'Product_ID':'Unique_Products'}, inplace=True)
sns.histplot(user_product_count['Unique_Products'], bins=20, kde=True)
```

Q15: Create a heatmap to analyze the correlation between numerical features in the dataset. Which features are most strongly correlated with the Purchase amount?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns
numeric_cols = ['Product_Category_1',
                'Product_Category_2',
                'Product_Category_3',
                'Amount']

corr = data[numeric_cols].corr()

plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Features")
plt.show()
```
